# Creating a singular KYC DF with all info

```python
import pandas as pd

# --- Copies ---
individuals = df_kyc_individual.copy()
occupations = df_kyc_occupation_codes.copy()
businesses = df_kyc_smallbusiness.copy()
industries = df_kyc_industry_codes.copy()
labels = df_labels.copy()
    
# --- Ensure consistent key data types ---
# Customer IDs
for df in [individuals, businesses, labels]:
    df["customer_id"] = df["customer_id"].astype("string")
# Occupation codes
individuals["occupation_code"] = individuals["occupation_code"].astype("string")
occupations["occupation_code"] = occupations["occupation_code"].astype("string")
# Industry codes
businesses["industry_code"] = businesses["industry_code"].astype("string")
industries["industry_code"] = industries["industry_code"].astype("string")
    
# --- Process individual customers ---
# Merge with occupation codes
individuals_enriched = individuals.merge(
    occupations[["occupation_code", "occupation_title"]],
    on="occupation_code",
    how="left"
)
# Mark as individual (not business)
individuals_enriched["is_small_business"] = 0
# Handle special occupation categories
SPECIAL_OCCUPATIONS = {"other", "retired", "self_employed", "student", "unemployed"}
# Normalize occupation code for matching
individuals_enriched["occupation_code_norm"] = (
    individuals_enriched["occupation_code"]
    .astype("string")
    .str.strip()
    .str.lower()
)
# Fill missing titles for special categories
mask = (
    individuals_enriched["occupation_title"].isna() & 
    individuals_enriched["occupation_code_norm"].isin(SPECIAL_OCCUPATIONS)
)
individuals_enriched.loc[mask, "occupation_title"] = individuals_enriched.loc[mask, "occupation_code_norm"]
# Clean up helper column
individuals_enriched = individuals_enriched.drop(columns=["occupation_code_norm"])

# --- Process small businesses ---
businesses_enriched = businesses.merge(
    industries[["industry_code", "industry"]],
    on="industry_code",
    how="left",
    validate="m:1"  # Many businesses to one industry
)
# Mark as business
businesses_enriched["is_small_business"] = 1
    
# --- Combine all customers ---
kyc_master = pd.concat(
    [individuals_enriched, businesses_enriched],
    axis=0,
    ignore_index=True
)

# --- Add customer labels ---
kyc_master = kyc_master.merge(labels, on="customer_id", how="left")

print("kyc_master shape:", kyc_master.shape)
print("unique customer_id:", kyc_master["customer_id"].nunique())
print("labeled rows:", kyc_master["label"].notna().sum())
print("label value counts (including NaN):")
print("Master KYC DF columns", kyc_master.columns)
print(kyc_master["label"].value_counts(dropna=False))

# display
kyc_master.head()

# save
kyc_master.to_csv(os.path.join(PROCESSED_DIR, "kyc_master.csv"), index=False)
```

```python
# =============================================================================
# Utils
# =============================================================================

def coerce_datetime(date):
    """
    Convert series to naive UTC datetime.
    
    Invalid values become NaT. All times are normalized to UTC
    but returned as timezone-naive datetime objects.
    """
    return pd.to_datetime(date, errors="coerce", utc=True).dt.tz_convert(None)


def safe_div(numerator, denominator, fill=0.0):
    """
    Safely divide numerator by denominator, handling division by zero and invalid values.
    Works for scalars and pandas Series. Returns the same type as input.
    
    Args:
        numerator: Scalar or Series
        denominator: Scalar or Series
        fill_value: Value to use when division is invalid
        
    Returns:
        Result of division with invalid values replaced by fill_value
    """
    try:
        result = numerator / denominator
    except ZeroDivisionError:
        return fill
    # Handle invalid values for scalars
    if np.isscalar(result):
        if pd.isna(result) or np.isinf(result):
            return fill
        return result
    # For Series/arrays
    return result.replace([np.inf, -np.inf, np.nan], fill)


def most_frequent_value(series: pd.Series):
    """
    Return the most frequent value in a Series.
    
    Handles ties deterministically by returning the smallest value.
    Returns NaN for empty or all-NaN Series.
    """
    if series.empty or series.dropna().empty:
        return np.nan
    # Get frequencies of non-null values
    counts = series.dropna().value_counts()
    # Find maximum frequency
    max_count = counts.max()
    # Get all values with maximum frequency, sorted for deterministic tie-breaking
    top_values = sorted(counts[counts == max_count].index)
    return top_values[0]


def parse_money(series: pd.Series):
    """
    Parse monetary values from strings into numeric format.
    
    Handles:
    - Plain numbers: "50000"
    - Currency symbols: "$50,000", "500"
    - Thousands separators: "50,000.50"
    - Whitespace: " 75000 "
    - Negative amounts: "-$5,000"
    
    Args:
        series: Series containing money values as strings

    Returns:
        Series of float values with invalid entries as NaN
    """
    if series.empty:
        return pd.Series([], dtype=float)
    # Convert to string type and strip whitespace
    s = series.astype("string").str.strip()
    # Remove currency symbols and thousands separators
    s = s.str.replace(r"[$,]", "", regex=True)
    # Convert to numeric, coercing errors to NaN
    return pd.to_numeric(s, errors="coerce")
```

```python
# =============================================================================
# 1) Per-channel amount aggregates
# =============================================================================

def aggregate_channel_amounts(df: pd.DataFrame, channel_name: str) -> pd.DataFrame:
    """
    Calculate transaction aggregates for a payment channel.

    Args:
        df: DataFrame with columns 'customer_id' and 'amount_cad'
        channel_name: Channel identifier for column naming

    Returns:
        DataFrame with additional columns: 
        n_txn_{ch}, sum_amt_{ch}, mean_amt_{ch}, median_amt_{ch}, std_amt_{ch}, cv_amt_{ch}, has_{ch}
    """
    # group and aggregate
    agg = (
        df.groupby("customer_id")["amount_cad"]
          .agg(n_txn="count", sum_amt="sum", mean_amt="mean", median_amt="median", std_amt="std")
          .reset_index()
    )
    # rename columns with channel prefix
    agg = agg.rename(columns={
        "n_txn": f"n_txn_{channel_name}",
        "sum_amt": f"sum_amt_{channel_name}",
        "mean_amt": f"mean_amt_{channel_name}",
        "median_amt": f"median_amt_{channel_name}",
        "std_amt": f"std_amt_{channel_name}",
    })

    # std NaN when n_txn==1 -> 0
    agg[f"std_amt_{channel_name}"] = agg[f"std_amt_{channel_name}"].fillna(0.0)

    # add coefficient of variation metric
    mean_col = f"mean_amt_{channel_name}"
    std_col  = f"std_amt_{channel_name}"
    cv_col   = f"cv_amt_{channel_name}"
    # calculate CV = std / mean, handle edge cases
    cv = agg[std_col] / agg[mean_col].replace(0, np.nan)
    # replace infinite values and NaNs with 0
    agg[cv_col] = cv.replace([np.inf, -np.inf], np.nan).fillna(0.0)

    # add indicator column
    agg[f"has_{channel_name}"] = (agg[f"n_txn_{channel_name}"] > 0).astype(int)
    return agg


# =============================================================================
# 2) Build txn_all (stack all channels) with column names
# =============================================================================

def make_txn_all(channel_dfs: dict) -> pd.DataFrame:
    """
    Combine transaction data from multiple channels into standardized format.

    Args:
        channel_dfs: Dictionary mapping channel names to DataFrames.
            Example: {"abm": df_abm, "card": df_card, "eft": df_eft}
            
            Required (present in ALL channels):
            transaction_id, customer_id, amount_cad, debit_credit, transaction_datetime    

            Optional by channel:
            - ABM + Card: country, province, city
            - ABM only: cash_indicator
            - Card only: merchant_category, ecommerce_ind
    
    Returns:
        Standardized DataFrame with transactions from all channels.

        Output columns (standardized):
        transaction_id, customer_id, amount_cad, dc_flag, txn_ts, channel,
        country, province, city, cash_indicator, merchant_category, ecommerce_ind
    """
    # Define schema constants
    REQUIRED_COLUMNS = {
        "transaction_id", "customer_id", "amount_cad", "dc_flag", "txn_ts"
    }
    OPTIONAL_COLUMNS = [
        "country", "province", "city", 
        "cash_indicator", "merchant_category", "ecommerce_ind"
    ]
    OUTPUT_COLUMNS = [
        "transaction_id", "customer_id", "amount_cad", "dc_flag", 
        "txn_ts", "channel", "country", "province", "city", 
        "cash_indicator", "merchant_category", "ecommerce_ind"
    ]

    # process each channel
    parts = []
    for ch, df in channel_dfs.items():
        if df is None or df.empty:
            continue
        
        # add channel name column
        tmp = df.copy()
        tmp["channel"] = ch
        tmp = tmp.rename(columns={
            "debit_credit": "dc_flag",
            "transaction_datetime": "txn_ts",
        })

        # ensure required columns exist
        missing = REQUIRED_COLUMNS - set(tmp.columns)
        if missing:
            raise ValueError(f"Channel '{ch}' is missing required columns: {missing}")

        # normalize fields
        tmp["txn_ts"] = coerce_datetime(tmp["txn_ts"])
        tmp["dc_flag"] = tmp["dc_flag"].astype(str).str.upper().str.strip()

        # add optional columns if absent in that channel
        for col in OPTIONAL_COLUMNS:
            if col not in tmp.columns:
                tmp[col] = np.nan
        
        parts.append(tmp[OUTPUT_COLUMNS])

    # concatenate channels together into one dataframe
    if not parts:
        return pd.DataFrame(columns=OUTPUT_COLUMNS)

    return pd.concat(parts, ignore_index=True)


# =============================================================================
# 3) KYC feature block (individual + smallbusiness unified)
# =============================================================================

def build_kyc_master(
    kyc_individual: pd.DataFrame,
    kyc_smallbusiness: pd.DataFrame,
    kyc_industry_codes: pd.DataFrame | None = None,
    kyc_occupation_codes: pd.DataFrame | None = None,
) -> pd.DataFrame:
    """
    Build unified customer KYC master table from individual and business sources.

    Args:
        kyc_individual: Individual customer KYC data
        kyc_smallbusiness: Small business customer KYC data  
        kyc_industry_codes: Industry code lookup table
        kyc_occupation_codes: Occupation code lookup table
    
    Returns:
        Unified DataFrame with customer attributes, one row for each customer,
        preferring business data for location fields when available.
        Fixes dtype issues by coercing join keys to the SAME dtype (string).
    """
    # copies of individual and business kyc dataframes
    ind = kyc_individual.copy()
    sb  = kyc_smallbusiness.copy()

    # coerce dates
    for c in ["birth_date", "onboard_date"]:
        if c in ind.columns:
            ind[c] = coerce_datetime(ind[c])
    for c in ["established_date", "onboard_date"]:
        if c in sb.columns:
            sb[c] = coerce_datetime(sb[c])

    # enrich kyc_individual with occupation titles
    # normalize join key dtypes
    if "occupation_code" in ind.columns:
        ind["occupation_code"] = ind["occupation_code"].astype("string").str.strip()
    if kyc_occupation_codes is not None and "occupation_code" in kyc_occupation_codes.columns:
        # normalize join key dtypes
        occ = kyc_occupation_codes.copy()
        occ["occupation_code"] = occ["occupation_code"].astype("string").str.strip()    
        # merge kyc_individual and kyc_occupation_codes
        if {"occupation_code", "occupation_title"}.issubset(occ.columns):
            ind = ind.merge(
                occ[["occupation_code", "occupation_title"]], on="occupation_code", how="left"
            )

    # enrich kyc_smallbusiness with industries
    # normalize join key dtypes
    if "industry_code" in sb.columns:
        sb["industry_code"] = sb["industry_code"].astype("string").str.strip()
    if kyc_industry_codes is not None and "industry_code" in kyc_industry_codes.columns:
        # normalize join key dtypes
        indc = kyc_industry_codes.copy()
        indc["industry_code"] = indc["industry_code"].astype("string").str.strip()
        # merge kyc_smallbusiness and kyc_industry_codes
        if indc is not None and {"industry_code", "industry"}.issubset(indc.columns):
            sb = sb.merge(
                indc[["industry_code", "industry"]], on="industry_code", how="left"
            )
    
    # merge individual + smallbusiness on customer_id 
    kyc = pd.merge(ind, sb, on="customer_id", how="outer", suffixes=("_ind", "_sb"))

    # unify fields with precedence: prefer smallbusiness for location/country if present 
    for col in ["country", "province", "city", "onboard_date"]:
        col_sb = f"{col}_sb"
        col_ind = f"{col}_ind"
        if col_sb in kyc.columns or col_ind in kyc.columns:
            kyc[col] = kyc.get(col_sb).combine_first(kyc.get(col_ind))

    # marital + income + birth_date live in individual
    kyc["marital_status"] = kyc.get("marital_status_ind", np.nan)
    kyc["income"] = kyc.get("income_ind", np.nan)
    kyc["birth_date"] = kyc.get("birth_date_ind", pd.NaT)

    # occupation fields from individual
    kyc["occupation_code"] = kyc.get("occupation_code_ind", pd.NA)
    kyc["occupation_title"] = kyc.get("occupation_title", pd.NA)

    # business fields from smallbusiness
    kyc["industry_code"] = kyc.get("industry_code", pd.NA)
    kyc["industry"] = kyc.get("industry", pd.NA)
    kyc["employee_count"] = kyc.get("employee_count", np.nan)
    kyc["sales"] = kyc.get("sales", np.nan)
    kyc["established_date"] = kyc.get("established_date", pd.NaT)

    # clean output
    OUTPUT_COLUMNS = [
        "customer_id",
        "country", "province", "city",
        "gender_ind",
        "marital_status",
        "occupation_code", "occupation_title",
        "income", "birth_date", "onboard_date",
        "industry_code", "industry",
        "employee_count", "sales", "established_date",
    ]
    keep = [c for c in OUTPUT_COLUMNS if c in kyc.columns]
    kyc = kyc[keep].copy()

    if "gender_ind" in kyc.columns:
        kyc = kyc.rename(columns={"gender_ind": "gender"})

    return kyc


def compute_kyc_features(
    kyc_master: pd.DataFrame,
    asof_ts: pd.Timestamp,
    occupation_risk_list=None,
) -> pd.DataFrame:
    """
    Compute KYC-derived features for customer analysis.
    
    Args:
        kyc_master: Master KYC DataFrame with customer information
        asof_ts: Reference timestamp for age calculations
        occupation_risk_list: List of high-risk occupation titles

    Returns:
        DataFrame with computed KYC features:
        account_age_days, account_lifetime, age, occupation, marital_status,
        business_age, employee_count, sales, industry_code,
        income_missing_flag, occupation_risk_flag, occupation_income_mismatch
    """
    df = kyc_master.copy()

    # ensure datetime columns are properly typed
    for c in ["birth_date", "onboard_date", "established_date"]:
        if c in df.columns:
            df[c] = coerce_datetime(df[c])

    # account_age_days: days since onboard_date
    if "onboard_date" in df.columns:
        df["account_age_days"] = (asof_ts - df["onboard_date"]).dt.days
        df["account_lifetime"] = df["account_age_days"]  # TODO: no explicit close date in your KYC tables
    else:
        df["account_age_days"] = np.nan
        df["account_lifetime"] = np.nan

    # age: from birth_date
    if "birth_date" in df.columns:
        df["age"] = ((asof_ts - df["birth_date"]).dt.days / 365.25).round(2)
    else:
        df["age"] = np.nan

    # occupation + marital_status
    occ_title = df["occupation_title"].astype("string") if "occupation_title" in df.columns else pd.Series(pd.NA, index=df.index, dtype="string")
    occ_raw   = df["occupation"].astype("string")       if "occupation" in df.columns       else pd.Series(pd.NA, index=df.index, dtype="string")

    df["occupation"] = occ_title.fillna(occ_raw)

    df["marital_status"] = df["marital_status"].astype("string") if "marital_status" in df.columns else pd.Series([pd.NA]*len(df), dtype="string")

    # business_age
    if "established_date" in df.columns:
        df["business_age"] = (asof_ts - df["established_date"]).dt.days
    else:
        df["business_age"] = np.nan

    # employee_count, sales, industry_code
    df["employee_count"] = pd.to_numeric(df["employee_count"], errors="coerce") if "employee_count" in df.columns else np.nan
    df["sales"] = pd.to_numeric(df["sales"], errors="coerce") if "sales" in df.columns else np.nan
    df["industry_code"] = df["industry_code"].astype("string") if "industry_code" in df.columns else pd.Series([pd.NA]*len(df), dtype="string")

    # income flags
    if "income" in df.columns:
        income = pd.to_numeric(df["income"], errors="coerce")
        df["income_missing_flag"] = income.isna().astype(int)
    else:
        income = pd.Series(np.nan, index=df.index)
        df["income_missing_flag"] = 1

    # keep only columns we want from KYC block (plus customer_id)
    OUTPUT_COLUMNS = [
        "customer_id",
        "account_age_days", "account_lifetime", "age",
        "occupation", "marital_status",
        "business_age", "employee_count", "sales", "industry_code",
        "income_missing_flag", "occupation_risk_flag", "occupation_income_mismatch",
        "country",  # keep home country for cross_border baseline
        "income",   # keep for low_income_high_txn_ratio
    ]
    keep = [c for c in OUTPUT_COLUMNS if c in df.columns]
    return df[keep].copy()


# =============================================================================
# 4) Rolling/window features across ALL channels (txn_all)
# =============================================================================

def _filter_window(txn: pd.DataFrame, asof_ts: pd.Timestamp, window_days: int) -> pd.DataFrame:
    start = asof_ts - pd.Timedelta(days=window_days)
    return txn[(txn["txn_ts"] > start) & (txn["txn_ts"] <= asof_ts)].copy()


def compute_txn_features(txn_all: pd.DataFrame, asof_ts: pd.Timestamp, window_days: int = 7, 
                         round_to: float = 100.0, same_amount_decimals: int = 2) -> pd.DataFrame:
    """
    Compute N-day transaction features across all payment channels.
    
    Args:
        transactions: DataFrame with transaction data (all channels)
        kyc_home_country: Series mapping customer_id to home country
        asof_timestamp: Reference timestamp for N-day window
        window_days: Number of days to consider
        round_to: Amount multiple for "round amount" detection
        same_amount_decimals: Decimal places for "same amount" detection
    
    Returns:
        DataFrame with customer_id and computed N-day features
        Uses ALL channels for core stats.
        Uses Card-only fields where available:
        - merchant_category for merchant features
        - ecommerce_ind for ecommerce features
        Uses country only where available (ABM + Card):
        - unique countries / most frequent country
        - cross_border_ratio computed relative to home country from KYC
    """
    # clean transaction date and filter to specified window
    required_cols = ["customer_id", "txn_ts", "amount_cad"]
    txn = txn_all.dropna(subset=required_cols).copy()
    w = _filter_window(txn, asof_ts, window_days).copy()
    if w.empty:
        return pd.DataFrame(columns=[
            "customer_id",
            f"transaction_frequency_{window_days}d",
            f"transaction_volume_{window_days}d",
            f"transaction_size_mean_{window_days}d",
            f"transaction_size_median_{window_days}d",
            f"transaction_size_variance_{window_days}d",
            f"transaction_round_amount_frequency_{window_days}d",
            f"transaction_same_amount_frequency_{window_days}d"
        ])
    # core N-day stats across ALL channels 
    g = w.groupby("customer_id")["amount_cad"]
    out = g.agg(
        transaction_frequency="count",
        transaction_volume="sum",
        transaction_size_mean="mean",
        transaction_size_median="median",
        transaction_size_variance="var",
    ).reset_index()
    out["transaction_size_variance"] = out["transaction_size_variance"].fillna(0.0)
    out = out.rename(columns={
        "transaction_frequency": f"transaction_frequency_{window_days}d",
        "transaction_volume": f"transaction_volume_{window_days}d",
        "transaction_size_mean": f"transaction_size_mean_{window_days}d",
        "transaction_size_median": f"transaction_size_median_{window_days}d",
        "transaction_size_variance": f"transaction_size_variance_{window_days}d"
    })
    # rounded amount frequency
    w["_is_round"] = (np.mod(w["amount_cad"].abs(), round_to) == 0).astype(int)
    round_freq = w.groupby("customer_id")["_is_round"].mean().reset_index()
    round_freq = round_freq.rename(columns={
        "_is_round": f"transaction_round_amount_frequency_{window_days}d"
    })
    out = out.merge(round_freq, on="customer_id", how="left")
    # same amount frequency
    w["_amt_key"] = w["amount_cad"].round(same_amount_decimals)
    dup = w.groupby(["customer_id", "_amt_key"])["_amt_key"].transform("size") >= 2
    w["_is_same_amt"] = dup.astype(int)
    same_amt = w.groupby("customer_id")["_is_same_amt"].mean().reset_index()
    same_amt = same_amt.rename(columns={
        "_is_same_amt": f"transaction_same_amount_frequency_{window_days}d"
    })
    out = out.merge(same_amt, on="customer_id", how="left")
    return out


def compute_txn_channel_features(txn_all: pd.DataFrame, kyc_home_country: pd.Series, asof_ts: pd.Timestamp) -> pd.DataFrame:
    # clean transaction date and filter to specified window
    required_cols = ["customer_id", "txn_ts", "amount_cad"]
    txn = txn_all.dropna(subset=required_cols).copy()
    w = _filter_window(txn, asof_ts, 90).copy()
    if w.empty:
        return pd.DataFrame(columns=[
            "customer_id",
            "cross_border_ratio",
            "transaction_unique_countries",
            "transaction_most_frequent_country",
            "transaction_ecommerce_ratio",
            "transaction_ecommerce_frequency_ratio",
            "transaction_unique_merchants",
            "transaction_most_frequent_merchant",
        ])
    
    # merchant features (Card only: merchant_category) 
    w_merch = w.dropna(subset=["merchant_category"]).copy()
    if not w_merch.empty:
        w_merch["_m"] = w_merch["merchant_category"].astype("string")
        out = w_merch.groupby("customer_id")["_m"].agg(
            transaction_unique_merchants=pd.Series.nunique,
            transaction_most_frequent_merchant=most_frequent_value
        ).reset_index()
    else:
        out = pd.DataFrame({
            "customer_id": w["customer_id"],
            "transaction_unique_merchants": 0,
            "transaction_most_frequent_merchant": np.nan
        })
    
    # ecommerce features (Card only: ecommerce_ind)
    w_eco = w.dropna(subset=["ecommerce_ind"]).copy()
    if not w_eco.empty:
        # e-commerce frequency-based ratio
        w_eco["_eco"] = pd.to_numeric(w_eco["ecommerce_ind"], errors="coerce").fillna(0).astype(int)
        eco_cnt = w_eco.groupby("customer_id")["_eco"].mean().reset_index()
        eco_cnt = eco_cnt.rename(columns={"_eco": "transaction_ecommerce_frequency_ratio"})
        # abm amount-based ratio (vectorized)
        w_eco["_abs_amt"] = w_eco["amount_cad"].abs()
        eco_num = w_eco[w_eco["_eco"]==1].groupby("customer_id")["_abs_amt"].sum()
        eco_den = w_eco.groupby("customer_id")["_abs_amt"].sum()
        eco_amt = safe_div(eco_num, eco_den, fill=0.0).reset_index()
        eco_amt.columns = ["customer_id", "transaction_ecommerce_ratio"]
    else:
        eco_cnt = pd.DataFrame({"customer_id": out["customer_id"], "transaction_ecommerce_frequency_ratio": 0.0})
        eco_amt = pd.DataFrame({"customer_id": out["customer_id"], "transaction_ecommerce_ratio": 0.0})
    out = out.merge(eco_amt, on="customer_id", how="left").merge(eco_cnt, on="customer_id", how="left")

    # country features (ABM + Card only: country present) 
    w_ctry = w.dropna(subset=["country"]).copy()
    if not w_ctry.empty:
        w_ctry["_ctry"] = w_ctry["country"].astype("string")
        ctry = w_ctry.groupby("customer_id")["_ctry"].agg(
            transaction_unique_countries=pd.Series.nunique,
            transaction_most_frequent_country=most_frequent_value
        ).reset_index()
        # cross-border relative to home country from KYC
        w_ctry["home_country"] = w_ctry["customer_id"].map(kyc_home_country).astype("string")
        w_ctry["_is_xb"] = (
            w_ctry["home_country"].notna() & 
            (w_ctry["_ctry"].str.upper() != 
             w_ctry["home_country"].str.upper())
        ).astype(int)
        cbr = w_ctry.groupby("customer_id")["_is_xb"].mean()
        cbr = cbr.reset_index(name="cross_border_ratio")
    else:
        ctry = pd.DataFrame({
            "customer_id": out["customer_id"],
            "transaction_unique_countries": 0,
            "transaction_most_frequent_country": np.nan
        })
        cbr = pd.DataFrame({"customer_id": out["customer_id"], "cross_border_ratio": 0.0})
    out = out.merge(ctry, on="customer_id", how="left").merge(cbr, on="customer_id", how="left")
    # numeric fills
    num_cols = [c for c in out.columns if c != "customer_id" and pd.api.types.is_numeric_dtype(out[c])]
    out[num_cols] = out[num_cols].fillna(0.0)
    return out


def compute_max_daily_frequency(txn_all: pd.DataFrame, asof_ts: pd.Timestamp, window_days: int) -> pd.DataFrame:
    # clean transaction date and filter to specified window
    txn = txn_all.dropna(subset=["customer_id", "txn_ts"]).copy()
    w = _filter_window(txn, asof_ts, window_days).copy()
    if w.empty:
        return pd.DataFrame(columns=["customer_id", f"max_transaction_frequency_{window_days}d"])
    # calculate maximum number of transactions per day for each customer
    w["_date"] = w["txn_ts"].dt.date
    daily = w.groupby(["customer_id", "_date"]).size().reset_index(name="_n")
    mx = daily.groupby("customer_id")["_n"].max().reset_index()
    return mx.rename(columns={"_n": f"max_transaction_frequency_{window_days}d"})


def compute_days_since_last_transaction(txn_all: pd.DataFrame, asof_ts: pd.Timestamp) -> pd.DataFrame:
    # remove rows with missing essential data
    txn = txn_all.dropna(subset=["customer_id", "txn_ts"]).copy()
    # calculate days elapsed since last transaction
    last = txn.groupby("customer_id")["txn_ts"].max().reset_index(name="_last_ts")
    last["days_since_last_transaction"] = (asof_ts - last["_last_ts"]).dt.days
    return last[["customer_id", "days_since_last_transaction"]]


def compute_below_threshold_frequency(
    txn_all: pd.DataFrame,
    asof_ts: pd.Timestamp,
    threshold: float,
    lower_frac: float = 0.90,   # 0.90 => 9,0009,999 when threshold=10,000
) -> pd.DataFrame:
    # clean transaction date and filter to specified window
    txn = txn_all.dropna(subset=["customer_id", "txn_ts", "amount_cad"]).copy()
    w = _filter_window(txn, asof_ts, 90).copy()
    if w.empty:
        return pd.DataFrame(columns=["customer_id", "count_txn_below_threshold_frequency"])
    # prepare upper and lower thresholds
    lower = threshold * lower_frac
    # use absolute amount for threshold comparison
    amt = w["amount_cad"].abs()
    # "Near threshold" band: [lower, threshold)
    w["_near_threshold"] = ((amt >= lower) & (amt < threshold)).astype(int)
    # Frequency = share of txns in the band in the last 7 days
    out = (
        w.groupby("customer_id")["_near_threshold"]
            .mean()
            .reset_index(name="count_txn_below_threshold_frequency")
    )
    return out

# =============================================================================
# 5) Flow features using Debit/Credit (D = out, C = in)
# =============================================================================

def compute_incoming_volume(txn_all: pd.DataFrame, asof_ts: pd.Timestamp, window_days: int = 90) -> pd.DataFrame:
    """
    Calculate total incoming transaction volume over N days.

    Args:
        txn_all: DataFrame with columns: customer_id, txn_ts, amount_cad, dc_flag
        asof_ts: End timestamp of analysis window
        window_days: Number of days to consider
        
    Returns:
        DataFrame with columns:
        - customer_id: Customer identifier
        - incoming_volume_Nd: Total incoming amount in N days in CAD
    """
    # clean transaction date and filter to specified window: N days
    txn = txn_all.dropna(subset=["customer_id", "txn_ts", "amount_cad", "dc_flag"]).copy()
    w = _filter_window(txn, asof_ts, window_days)
    if w.empty:
        return pd.DataFrame(columns=["customer_id", f"incoming_volume_{window_days}d"])
    # filter for incoming transactions and sum for each customer
    inc = w[w["dc_flag"]=="C"]
    inc = inc.groupby("customer_id")["amount_cad"].sum()
    inc = inc.reset_index(name=f"incoming_volume_{window_days}d")
    return inc


def compute_channel_credit_debit_and_ratios(txn_all: pd.DataFrame, asof_ts: pd.Timestamp) -> pd.DataFrame:
    """
    Calculate channel-based transaction ratios and credit/debit volumes in 90 days.

    Args:
        txn_all: DataFrame with columns: customer_id, txn_ts, amount_cad, 
            channel, dc_flag (optional)
        asof_ts: End timestamp of analysis windo

    Returns:
        DataFrame with columns:
        - customer_id
        - channel_credit_volume (ABS sum of Credit C in last 90 days across all channels)
        - channel_debit_volume  (ABS sum of Debit  D in last 90 days across all channels)
        - cash_ratio            (ABS volume of ABM over ABS total volume, last 90 days)
        - wire_ratio            (ABS volume of Wire over ABS total volume, last 90 days)
    """
    # clean transaction date and filter to specified window: 90 days
    txn = txn_all.dropna(subset=["customer_id", "txn_ts", "amount_cad"]).copy()
    w = _filter_window(txn, asof_ts, 90).copy()
    if w.empty:
        return pd.DataFrame(columns=["customer_id", "channel_credit_volume", "channel_debit_volume", "cash_ratio", "wire_ratio"])
    # use absolute amount for computing transaction ratios
    w["_abs"] = w["amount_cad"].abs()
    # calculate total absolute volume per customer
    total_abs = w.groupby("customer_id")["_abs"].sum()
    # credit/debit volumes (require dc_flag)
    if "dc_flag" in w.columns:
        w2 = w.dropna(subset=["dc_flag"]).copy()
        credit = w2[w2["dc_flag"]=="C"].groupby("customer_id")["_abs"].sum()
        debit  = w2[w2["dc_flag"]=="D"].groupby("customer_id")["_abs"].sum()
    else:
        credit = pd.Series(dtype=float)
        debit  = pd.Series(dtype=float)
    # map volumes and ratios to customers
    out = pd.DataFrame({"customer_id": w["customer_id"].unique()})
    out["channel_credit_volume"] = out["customer_id"].map(credit).fillna(0.0)
    out["channel_debit_volume"]  = out["customer_id"].map(debit).fillna(0.0)
    # cash_ratio: ABM as "cash-like" (you can expand if needed)
    ch = w["channel"].astype(str).str.lower()
    abm_abs = w[ch=="abm"].groupby("customer_id")["_abs"].sum()
    wire_abs = w[ch=="wire"].groupby("customer_id")["_abs"].sum()
    # divide ABM/wire amounts by total amounts
    out["cash_ratio"] = out["customer_id"].map(safe_div(abm_abs, total_abs, fill=0.0)).fillna(0.0)
    out["wire_ratio"] = out["customer_id"].map(safe_div(wire_abs, total_abs, fill=0.0)).fillna(0.0)
    return out


def compute_abm_deposit_withdrawal(txn_all: pd.DataFrame, asof_ts: pd.Timestamp) -> pd.DataFrame:
    """
    Calculate ABM cash deposit and withdrawal volumes in 90 days.

    Args:
        txn_all: DataFrame with columns: customer_id, txn_ts, amount_cad, 
                     dc_flag, channel
        asof_ts: End timestamp of analysis window
        
    Returns:
        DataFrame with columns:
        - customer_id
        - abm_deposit_volume: Credit (C) volume in ABM last 90 days
        - abm_withdrawal_volume: Debit (D) volume in ABM last 90 days
    """
    # clean transaction date and filter to specified window: 90 days
    txn = txn_all.dropna(subset=["customer_id", "txn_ts", "amount_cad", "dc_flag", "channel"]).copy()
    w = _filter_window(txn, asof_ts, 90)
    w = w[w["channel"].astype(str).str.lower()=="abm"].copy()
    if w.empty:
        return pd.DataFrame(columns=["customer_id", "abm_cash_deposit_volume", "abm_cash_withdrawal_volume"])
    # use absolute amounts
    w["_abs"] = w["amount_cad"].abs()
    # calculate deposit and withdrawal volumes from ABM transactions
    dep = w[w["dc_flag"]=="C"].groupby("customer_id")["_abs"].sum()
    wd  = w[w["dc_flag"]=="D"].groupby("customer_id")["_abs"].sum()
    # map volumes to customers
    out = pd.DataFrame({"customer_id": w["customer_id"].unique()})
    out["abm_cash_deposit_volume"] = out["customer_id"].map(dep).fillna(0.0)
    out["abm_cash_withdrawal_volume"] = out["customer_id"].map(wd).fillna(0.0)
    return out


def compute_hold_time_features(txn_all: pd.DataFrame, asof_ts: pd.Timestamp) -> pd.DataFrame:
    """
    Calculate fund hold time features between incoming and outgoing transactions in 90 days.
    For each Credit (C), find the next Debit (D) and compute time delta (days).
    
    Args:
        txn_all: DataFrame with columns: customer_id, txn_ts, dc_flag
        asof_ts: End timestamp of analysis window

    Returns:
        DataFrame with columns:
        - customer_id
        - median_hold_time_funds: Median hours between incoming and next outgoing
        - incoming_to_outgoing_time_mean: Mean hours between incoming and next outgoing
    """
    # clean transaction date and filter to specified window: 90 days
    txn = txn_all.dropna(subset=["customer_id", "txn_ts", "dc_flag"]).copy()
    w = _filter_window(txn, asof_ts, 90).copy()
    if w.empty:
        return pd.DataFrame(columns=["customer_id", "median_hold_time_funds", "incoming_to_outgoing_time_mean"])
    # sort by customer and timestamp for efficient processing
    w = w.sort_values(["customer_id", "txn_ts"])
    # loop over each customer
    pairs = []
    for cid, d in w.groupby("customer_id", sort=False):
        in_times = d.loc[d["dc_flag"]=="C", "txn_ts"].tolist()
        out_times = d.loc[d["dc_flag"]=="D", "txn_ts"].tolist()
        # if no incoming or outgoing transactions, proceed to the next customer
        if not in_times or not out_times:
            continue
        # for each incoming transaction, find the next outgoing transaction
        out_idx = 0
        for t_in in in_times:
            # skip outgoing transactions that occurred before this incoming
            while out_idx < len(out_times) and out_times[out_idx] <= t_in:
                out_idx += 1
            # if there's a future outgoing transaction, calculate hold time
            if out_idx < len(out_times):
                delta_days = (out_times[out_idx] - t_in).total_seconds() / (3600)  # convert to hours
                pairs.append((cid, delta_days))
    # if hold times are empty, return default values for all customers
    if not pairs:
        base = pd.DataFrame({"customer_id": w["customer_id"].unique()})
        base["median_hold_time_funds"] = 0.0
        base["incoming_to_outgoing_time_mean"] = 0.0
        return base
    # calculate median and mean hold times for each customer
    p = pd.DataFrame(pairs, columns=["customer_id", "_hold_days"])
    out = p.groupby("customer_id")["_hold_days"].agg(
        median_hold_time_funds="median",
        incoming_to_outgoing_time_mean="mean"
    ).reset_index()
    return out


# =============================================================================
# 6) MASTER builder: ALL requested features + your original per-channel + shares
# =============================================================================

def build_all_features(
    *,
    # transactions
    df_abm: pd.DataFrame,
    df_card: pd.DataFrame,
    df_cheque: pd.DataFrame,
    df_eft: pd.DataFrame,
    df_emt: pd.DataFrame,
    df_wire: pd.DataFrame,
    df_wu: pd.DataFrame,
    # KYC tables
    kyc_individual: pd.DataFrame,
    kyc_smallbusiness: pd.DataFrame,
    kyc_industry_codes: pd.DataFrame,
    kyc_occupation_codes: pd.DataFrame,
    # config
    txn_threshold: float = 10000.0,
    occupation_risk_list=None,
    occupation_median_map: dict | pd.Series | None = None,  # industry_code -> median baseline
) -> pd.DataFrame:
    # Step 0: Prepare channel data
    channel_dfs = {
        "abm": df_abm,
        "card": df_card,
        "cheque": df_cheque,
        "eft": df_eft,
        "emt": df_emt,
        "wire": df_wire,
        "wu": df_wu,
    }
    # combine all transactions
    txn_all = make_txn_all(channel_dfs)
    # as-of time = max transaction timestamp
    asof_ts = txn_all["txn_ts"].max()
    if pd.isna(asof_ts):
        asof_ts = pd.Timestamp.utcnow().tz_localize(None)
    # KYC master + KYC features
    kyc_master = build_kyc_master(
        kyc_individual=kyc_individual,
        kyc_smallbusiness=kyc_smallbusiness,
        kyc_industry_codes=kyc_industry_codes,
        kyc_occupation_codes=kyc_occupation_codes,
    )
    kyc_feat = compute_kyc_features(
        kyc_master=kyc_master,
        asof_ts=asof_ts,
        occupation_risk_list=occupation_risk_list,
    )
    # home country baseline for cross-border
    if "country" in kyc_master.columns:
        home_country = kyc_master.set_index("customer_id")["country"].astype("string")
    else:
        home_country = pd.Series(dtype="string")
    # start feature table from KYC
    feat = kyc_feat.copy()

    # Step 1: Add per-channel aggregate features and merge
    for ch, df in channel_dfs.items():
        agg = aggregate_channel_amounts(df, ch)
        feat = feat.merge(agg, on="customer_id", how="left")
    # fill 0 where semantically correct
    n_cols   = [c for c in feat.columns if c.startswith("n_txn_")]
    sum_cols = [c for c in feat.columns if c.startswith("sum_amt_")]
    has_cols = [c for c in feat.columns if c.startswith("has_")]
    std_cols = [c for c in feat.columns if c.startswith("std_amt_")]
    cv_cols  = [c for c in feat.columns if c.startswith("cv_amt_")]
    for cols in [n_cols, sum_cols, has_cols, std_cols, cv_cols]:
        if cols:
            feat[cols] = feat[cols].fillna(0.0)
    # do NOT fill mean/median after merge (keeps NaN = no usage)

    # Step 2: Add cross-channel totals/shares
    feat["sum_amt_total"] = feat[sum_cols].sum(axis=1) if sum_cols else 0.0
    feat["n_txn_total"]   = feat[n_cols].sum(axis=1) if n_cols else 0.0
    feat["num_channels_used"] = feat[has_cols].sum(axis=1) if has_cols else 0.0
    # calculate channel shares
    for c in sum_cols:
        share_col = c.replace("sum_amt_", "share_amt_")
        feat[share_col] = safe_div(feat[c], feat["sum_amt_total"], fill=0.0)
    # calculate maximum channel share
    share_cols = [c for c in feat.columns if c.startswith("share_amt_")]
    feat["max_channel_share"] = feat[share_cols].max(axis=1) if share_cols else 0.0

    # Step 3: Add 7-day/90-day features (with correct only card has merchant/ecomm, only abm+card has country)
    f7 = compute_txn_features(txn_all, asof_ts, window_days=7)
    f90 = compute_txn_features(txn_all, asof_ts, window_days=90)
    f = compute_txn_channel_features(txn_all, home_country, asof_ts)
    feat = feat.merge(f7, on="customer_id", how="left").merge(f90, on="customer_id", how="left")
    feat = feat.merge(f, on="customer_id", how="left")
    
    # Step 4: Add frequency features
    # max transaction frequency 7 days / 90 days
    mx7 = compute_max_daily_frequency(txn_all, asof_ts, window_days=7)
    mx90 = compute_max_daily_frequency(txn_all, asof_ts, window_days=90)
    feat = feat.merge(mx7, on="customer_id", how="left").merge(mx90, on="customer_id", how="left")
    
    # Step 5: Add recency features
    rec = compute_days_since_last_transaction(txn_all, asof_ts)
    feat = feat.merge(rec, on="customer_id", how="left")
    
    # Step 6: Add incoming volume (7 days / 90 days, Credit=C)
    inc7 = compute_incoming_volume(txn_all, asof_ts, window_days=7)
    inc90 = compute_incoming_volume(txn_all, asof_ts, window_days=90)
    feat = feat.merge(inc7, on="customer_id", how="left").merge(inc90, on="customer_id", how="left")
    
    # Step 7: Add below-threshold frequency (90 days)
    below = compute_below_threshold_frequency(txn_all, asof_ts, threshold=txn_threshold)
    feat = feat.merge(below, on="customer_id", how="left")

    # Step 8: credit/debit volumes + cash/wire ratios (90 days)
    flow = compute_channel_credit_debit_and_ratios(txn_all, asof_ts)
    feat = feat.merge(flow, on="customer_id", how="left")

    # Step 9: Add ABM deposit/withdraw volumes (90 days, using C vs D)
    abm = compute_abm_deposit_withdrawal(txn_all, asof_ts)
    feat = feat.merge(abm, on="customer_id", how="left")

    # Step 10: Add hold time features (90 days)
    hold = compute_hold_time_features(txn_all, asof_ts)
    feat = feat.merge(hold, on="customer_id", how="left")

    # Step 11: Finalize features
    # fill numeric NaNs (no activity in windows => 0)
    DO_NOT_ZERO = {"income", "sales", "employee_count", "capacity_amount"}  # keep unknowns as NaN, not 0
    numeric_cols = [
        c for c in feat.columns
        if c != "customer_id"
        and pd.api.types.is_numeric_dtype(feat[c])
        and c not in DO_NOT_ZERO
    ]
    if numeric_cols:
        feat[numeric_cols] = feat[numeric_cols].fillna(0.0)
    # txn_to_sales_ratio: 90 days, businesses only, but this naturally becomes NaN for individuals)
    sales = pd.to_numeric(feat.get("sales", np.nan), errors="coerce")
    txn_vol = pd.to_numeric(feat["transaction_volume_90d"], errors="coerce").abs()
    feat["txn_to_sales_ratio"] = (txn_vol / sales).replace([np.inf, -np.inf], np.nan)

    return feat
```

# Add engineered features to kyc_master df from above

```python
# Step 0: Prepare data
# Ensure consistent key dtype
kyc_master["customer_id"] = kyc_master["customer_id"].astype("string").str.strip()
data["customer_id"] = data["customer_id"].astype("string").str.strip()

# Only bring engineered columns that don't already exist in kyc_master
new_cols = [c for c in data.columns if c not in kyc_master.columns]

kyc_plus = kyc_master.merge(
    data[["customer_id"] + new_cols],
    on="customer_id",
    how="left"
)

# Step 1: Add age
kyc_plus["birth_date"] = pd.to_datetime(kyc_plus["birth_date"], errors="coerce")
asof_ts = pd.Timestamp.today().normalize()
kyc_plus["age"] = (
    (asof_ts - kyc_plus["birth_date"])
    .dt.days / 365.25
).round(0).astype("Int64")

# Step 2: Add txn_volume_vs_income (90 days; individuals only; income exists only for individuals)
# Parse income safely (handles "$50,000" and "50000")
income_num = pd.to_numeric(
    kyc_plus["income"].astype("string").str.replace(r"[\$,]", "", regex=True).str.strip(),
    errors="coerce"
)
valid = income_num > 0
txn_vol_90d = pd.to_numeric(kyc_plus.get("transaction_volume_90d", np.nan), errors="coerce").abs()
kyc_plus["txn_volume_vs_income"] = np.nan
kyc_plus.loc[valid, "txn_volume_vs_income"] = (txn_vol_90d.loc[valid] / income_num.loc[valid]).replace([np.inf, -np.inf], np.nan)

# Optional: income missing flag (useful)
kyc_plus["income_missing_flag"] = (income_num.isna() | (income_num <= 0)).astype(int)

# Step 3: Add txn_volume_vs_occupation_median (90 days)
# Expect: occupation_median_map maps occupation string -> median income (float)
# We will use occupation_title 
occ_source_col = "occupation_title" if "occupation_title" in kyc_plus.columns else "occupation"
kyc_plus["txn_volume_vs_occupation_median"] = np.nan

if (occ_source_col in kyc_plus.columns) and ("transaction_volume_90d" in kyc_plus.columns):
    # normalize occupation values
    occ_key = (
        kyc_plus[occ_source_col]
        .astype("string")
        .str.strip()
        .str.lower()
    ).mask(lambda s: s.isna() | (s == ""), pd.NA)
    # normalize map keys once
    occ_map = {str(k).strip().lower(): float(v) for k, v in occupation_median_map.items()}
    occ_med = pd.to_numeric(occ_key.map(occ_map), errors="coerce")
    valid = occ_med > 0
    # compute ratio of txn_volume to occupation_median
    txn_vol_90d = pd.to_numeric(kyc_plus["transaction_volume_90d"], errors="coerce").abs()
    kyc_plus["txn_volume_vs_occupation_median"] = np.nan
    kyc_plus.loc[valid, "txn_volume_vs_occupation_median"] = txn_vol_90d.loc[valid] / occ_med.loc[valid]

print("kyc_master:", kyc_master.shape)
print("data:", data.shape)
print("kyc_plus:", kyc_plus.shape)
```
